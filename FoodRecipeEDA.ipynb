```python
# To analyze columns cook_time and prep_time as numeric we can convert them with the following code:
df_recipe['cook_time'] = df_recipe['cook_time'].str.replace(' M', '').fillna(0).astype(int)
df_recipe['prep_time'] = df_recipe['prep_time'].str.replace(' M', '').fillna(0).astype(int)
```

```python
# Expanding the nested ingredients column
df_recipe['ingredients'].str.split('|', expand=True)
```

```python
# split the column on separator "|", expand=True - will expand data to DataFrame
# melt() - transform DataFrame columns to rows - pairs of variable and value
# count the values by .value_counts()
df_recipe['ingredients'].str.split('|', expand=True).melt()['value'].value_counts()
```

#### Figuring out how to deal with missing values.
* df.isna() - returns new DataFrame with True/False values depending on the missing values in a cell* 
cmap = 'Greens' - is a color map.

```python
sns.heatmap(df_recipe.isna(),cmap = 'Greens');
```

#### The green horizontal lines are the missing data in the dataset, particulary in the cuisine, course, and the diet columns.

```python
# Select only numeric columns before applying .corr()
df_corr = df_recipe.select_dtypes(include=['number']).corr()
```

```python
# Darker color represent positive correlation: lighter - negative
# To get values use annot=True
plt.figure(figsize=(10, 6))
sns.heatmap(df_corr, cmap='Greens', annot=True)
plt.show()
```

<b> There is basically no correlation between the numerical value columns in this dataset. </b>

```python
cols = ['vote_count', 'rating',  'prep_time', 'cook_time']
plt.figure(figsize=(20, 5))
sns.boxplot(data=df_recipe[cols], orient='h');
```

```python
# Boxplot for just the first 18 results.
cols = ['prep_time', 'cook_time']
plt.figure(figsize=(20, 5))
sns.boxplot(data=df_recipe.head(18)[cols], orient='h');
```

#### Boxplot overview: 
* Boxplot has 3 main components:
* box (Interquartile Range - IQR) q1 - 25%, q2 - 50% or median, q3 - 75%.
* whiskers - min & max.
* outliers - points beyond the whiskers.

<b> We can notice that sometimes the whiskers don't match the min/max. They are calculated based on formulas. Upper and low whiskers are calculated as 1.5 multiplied on the interquartile range - IQR(the box). </b>
